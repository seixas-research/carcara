```python
from carcara import __version__
from carcara.core.trainer import Trainer
import warnings
warnings.filterwarnings("ignore")
import os

print(f"Carcará version = {__version__}")
```

```python
trainer = Trainer(name           = "model_test",
                  train_file     = "data/splits/train_seed_876.xyz",  # 400 configurations of Au bulk, 108 atoms each
                  valid_file     = "data/splits/valid_seed_876.xyz",  # 100 configurations of Au bulk, 108 atoms each
                  E0s            = {79: 3.800},                       # Reference energy for Au, in eV/atom
                  max_num_epochs = 200,
                  start_swa      = 100,
                  eval_interval  = 5,
                  num_channels   = 16,
                  max_L          = 0)
```

```python
trainer.run_train()
```

```python
from carcara.atoms_noise import AtomsNoiseGenerator
from ase.build import bulk
from ase.calculators.emt import EMT
```

```python
atoms = bulk("Au", "fcc", a=4.08, cubic=True).repeat((3,3,3))
calculator = EMT()
```

```python
generator = AtomsNoiseGenerator(atoms=atoms, calculator=calculator, noise_type='normal', seed=42)
generator.relax_structure(fmax=0.01, relax_cell=True)
generator.generate_samples(num_samples=100, noise_type='uniform', noise_level_pos=0.2, noise_level_cell=0.2, cell_mode='all')
generator.save_to_xyz("noise_samples.xyz", compute_ref=True)
```
